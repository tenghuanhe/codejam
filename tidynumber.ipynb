```python
def i2cs(n):
    return list(str(n))
```

```python
def isTidyNumber(n):
    cs = i2cs(n)
    for i in range(len(cs) - 1):
        if cs[i] > cs[i + 1]:
            return False
    return True
```

```python
def maxTidyNumber(n):
    if isTidyNumber(n):
        return n
    cs = i2cs(n)
    flag = 0
    for i in range(len(cs) - 1):
        if cs[i + 1] < cs[i]:
            flag = i
            break
    while i > 0:
        if cs[i] == cs[i - 1]:
            i -= 1
        else:
            break
    return int(''.join(cs[:i]) + str(int(cs[i]) - 1) + '9' * (len(cs) - i - 1))
```

```python
def sol(inputFile):
    res = ''
    with open(inputFile) as f:
        lineNumber = int(f.readline())
        for i in range(lineNumber):
            n = int(f.readline())
            res += 'Case #' + str(i + 1) + ': ' + str(maxTidyNumber(n)) + '\n'
    with open(inputFile.replace('.in', '.out'), 'w') as fo:
        fo.write(res)
    return res
```

```python
print sol('B-small-practice.in')
```
