```python
def getLevelPositionCurrsum(n):
    c = 0
    total = 0
    i = 1;
    while total <= n:
        total += i
        if total >= n:
            return (c, n + i - total, total)
        c += 1
        i *= 2
        
    return (0, 0, 0)

for i in range(1, 11):
    print i, getLevelPositionCurrsum(i)
```

```python
def solution(n, k):
    (l, p, cs) = getLevelPositionCurrsum(k)
    stalls = n - cs
    c = 2 ** l
    if stalls % c == 0:
        lrs = stalls / c
        if lrs % 2 == 0:
            return (lrs / 2, lrs / 2)
        else:
            return ((lrs + 1) / 2, (lrs - 1) / 2)
    else:
        re = stalls % c
        lrs = stalls / c
        if lrs % 2 == 0:
            if p <= re:
                return (lrs / 2 + 1, lrs / 2)
            else:
                return (lrs / 2, lrs / 2)
        else:
            if p <= re:
                return ((lrs + 1) / 2, (lrs + 1) / 2)
            else:
                return ((lrs + 1) / 2, (lrs - 1) / 2)
    
    
    
```

```python
def sol(inputFile):
    res = ''
    with open(inputFile) as f:
        lineNumber = int(f.readline())
        for i in range(lineNumber):
            line = f.readline()
            n = int(line.split(' ')[0])
            k = int(line.split(' ')[1])
            (a, b) = solution(n, k)
            res += 'Case #' + str(i + 1) + ': ' + str(a) + ' ' + str(b) + '\n'
    with open(inputFile.replace('.in', '.out'), 'w') as fo:
        fo.write(res)
    return res
```
